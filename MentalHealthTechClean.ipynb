```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/survey.csv")
```

```python
df = df.drop(columns = "Timestamp")
```

```python
df1 = df[df["Country"] == "United States"]
df1
```

```python
df = df1.drop(columns = "self_employed")
df = df.drop(columns = "no_employees")
df = df.drop(columns = "leave")
df = df.drop(columns = "comments")
df = df.drop(columns = "state")
df = df.drop(columns = "Country")
df = df.drop(columns = "coworkers")
df = df.drop(columns = "Gender")
```

```python
df.head()
```

```python
for col in df.columns:
 df.loc[df[col] == "No", col] = 0
 df.loc[df[col] == "Yes", col] = 1
 df.loc[df[col] == "Don't know", col] = 77
 df.loc[df[col] == "Not sure", col] = 77 
 df.loc[df[col] == "Often", col] = 5
 df.loc[df[col] == "Rarely", col] = 2
 df.loc[df[col] == "Sometimes", col] = 4
 df.loc[df[col] == "Never", col] = 1 
 df.loc[df[col] == "Maybe", col] = 33 
 df.loc[df[col] == "Some of them", col] = 3 
```

```python
df['work_interfere'] = df['work_interfere'].fillna(77)
```
